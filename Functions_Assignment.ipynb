#Q-6) Explain the concept of generators in Python and how they are defined?

#Ans-6) Generator is funcion of python which creates iterators using function & yield keyword. It creates iterators for each character or data in any element and hence does not gives the whole result in one go. It iterates one by one using the next() function using the indexes of elements and hence consumes less time and memory to work.

```python
#Example:
def square_no(n):
  for i in range(n):
    yield i**2


for i in square_no(4):
  print(i)
```

#Q-10) What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

2.   **Reduce()**- This functions is imported from functool first for bringing it into use for programming. It applies a function cumulatively to reduce the resultant size of the sequence into single value.

#Q-5) Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
#Ans-5)
class fibonacciIterator:
  def __init__(self,limit):
    self.limit=limit
    self.a=0
    self.b=1

  def __iter__(self):
    return self

  def __next__(self):
    fib_num=self.a
    self.a,self.b=self.b,self.a+self.b
    if fib_num>self.limit:
      raise StopIteration
    return fib_num



fib_iter=fibonacciIterator(12)

for num in fib_iter:
  print(num)
```

#Q-7) Implement a generator function that reads a file line by line and yields each line as a string

```python
#Ans-7)
def read_file_line_by_line(file_path):
    with open(file_path,'r') as file:
      for line in file:
        yield line.strip()
    except FileNotFoundError:
      print(f"Error: File is not found at path: {file_path}")
      return

file_path='shashank.txt'
for line in read_file_line_by_line(file_path):
  print(line)
```

#Q-10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
#Ans-10)
def remove_vowels_filter(text):
  vowels = set('aeiouAEIOU')  # Define a set of vowels
  filtered_chars = filter(lambda char: char not in vowels, text)
  return ''.join(filtered_chars)

input_string= input("Enter a String:")
output_string= remove_vowels_filter(input_string)
print("String without vowels:",output_string)

```
